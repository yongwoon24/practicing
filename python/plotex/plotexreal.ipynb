```python
import numpy as np
import pandas as pd
#https://jumin.mois.go.kr/ageStatMonth.do#none  ==> 데이터 다운받은 사이트
file_name = '201406_201406_연령별인구현황_월간.xlsx'
df_m=pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,E:Y')
df_m.index.values #인덱스의 구성과 인덱스 명이 어떻게 되어있는지 특수문자나 띄어쓰기를 확인
#df_m.loc['전국  ']
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int)
# df_m.iloc[0]
# type(df_m.iloc[0,1]) =>숫자(int)로 바뀌었는지 확인

```

```python
df_w=pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df_w.index.values
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)
df_w.iloc[0]
#df_w.index.values
```

```python
df_w.columns = df_m.columns
df_w.columns
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'Malgun Gothic' # AppleGothic ==> 맥에서는 맑은 고딕대신 사용
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
plt.figure(figsize=(10,5))
plt.barh(df_w.columns,df_w.iloc[0],label='여',color='lightblue')
plt.barh(df_m.columns,-df_m.iloc[0],label='남',color='coral')
plt.legend()
plt.title('2014년 인구 현황',size=20)
plt.savefig('2014년 인구 현황.png',dpi=1000)
```
