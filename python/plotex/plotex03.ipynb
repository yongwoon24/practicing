```python
import matplotlib.pyplot as plt

# 예시 데이터
schools = ['School A', 'School B', 'School C', 'School D']
student_counts = [350, 450, 300, 400]

# 파이 그래프 생성
plt.figure(figsize=(10, 7))
plt.pie(student_counts, labels=schools, autopct='%1.1f%%', startangle=140)

# 그래프 제목 추가
plt.title('Number of Students per School')

# 그래프 표시
plt.show()
```

```python
import pandas as pd

data = pd.read_excel('score.xlsx', index_col="지원번호")
```

```python
import matplotlib as mpl

import matplotlib.font_manager as fm
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.size'] = 10
df = pd.DataFrame(data)

school_counts = df.groupby('학교').size()
plt.figure(figsize=(10, 7))
plt.pie(school_counts, labels=school_counts.index, autopct='%1.1f%%', startangle=140)

plt.title('학교별 통계')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['이름'], df['국어'], color='skyblue')
plt.xlabel('이름')
plt.ylabel('국어 점수')
plt.title('학생당 국어 점수')
plt.show()

subjects = ['국어', '영어', '수학', '과학', '사회']
plt.figure(figsize=(14, 8))

for i in range(len(df)):
    plt.plot(subjects, df.loc[i, subjects], marker='o', label=df['이름'][i])

plt.xlabel('과목')
plt.ylabel('점수')
plt.title('학생당 과목별 점수')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# 첫 번째 서브플롯: 학생당 국어 점수 막대 그래프
axes[0].bar(df['이름'], df['국어'], color='skyblue')
axes[0].set_xlabel('이름')
axes[0].set_ylabel('국어 점수')
axes[0].set_title('학생당 국어 점수')

# 두 번째 서브플롯: 학생당 과목별 점수 꺾은선 그래프 (국어, 영어, 수학)
subjects = ['국어', '영어', '수학']
for subject in subjects:
    axes[1].plot(df['이름'], df[subject], marker='o', label=subject)

axes[1].set_xlabel('이름')
axes[1].set_ylabel('점수')
axes[1].set_title('학생당 과목별 점수 (국어, 영어, 수학)')
axes[1].legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```
